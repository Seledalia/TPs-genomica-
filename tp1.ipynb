En este ejercicio utilizaremos dos herramientas para ejemplificar la manipulación de archivos de texto en Unix, aprender a guardar resultados en archivos y concatenar comandos.


**Awk** es un lenguaje de procesamiento de texto y un comando diseñado para buscar, filtrar y manipular datos en archivos de texto, especialmente tablas o datos separados por delimitadores (como espacios o comas). Permite seleccionar líneas según patrones, extraer columnas, realizar cálculos y generar reportes.

**Grep** es un comando utilizado para buscar líneas que coincidan con un patrón de texto dentro de archivos o flujos de datos. Permite filtrar información rápidamente usando expresiones regulares y opciones como mostrar el número de línea, ignorar mayúsculas/minúsculas, o buscar recursivamente en directorios.


Al igual que muchos programas de Unix, podemos entender los parámetros posibles utilizando un parámetro específico que imprime más información. Para awk este parámetro es **-W help**, pero en la mayoría de los programas es **--help**. Para más información, recurrir a guías en internet, por [ejemplo](https://www.google.com/url?sa=t&source=web&rct=j&opi=89978449&url=https://www.geeksforgeeks.org/linux-unix/awk-command-unixlinux-examples/&ved=2ahUKEwiC_MzmioiQAxWeBLkGHS51Dh8QFnoECBkQAQ&usg=AOvVaw3tvNrjj9AJqSclSDOodvpg).


```python
!pwd
```

```python
%cd /content/sample_data/

# Me interesan únicamente las columnas "longitude" y "latitude" que están en la columna 1 y 2 del archivo california_housing_test
# Ver primeras lineas
!head -2 california_housing_test.csv
```

```python
# Imprimir únicamente las columnas de interés
!awk -F"," '{print $1,$2}' california_housing_test.csv
```

Desglocemos este comando:  

* **-F ","** *indica que las columnas se encuentran separadas por coma.*
* **'{print .... '**  *indica al programa a imprimir el resultado.*
* **\$1, \$2**    *indica al programa imprimir únicamente las columnas 1 y 2.*


```python
!awk -W help
```

Para el próximo ejercicio, sólo nos interesa el dato de la población en las coordenadas "-118.140000,33.880000", para encontrarlo entre las 3001 líneas posibles, utilizaremos el comando grep.

```python
# Encontrar líneas que presentan las coordenadas de interés

!grep -- "-118.140000,33.880000" california_housing_test.csv
```

```python
# Aislar el dato de Población
!grep -- "-118.140000,33.880000" california_housing_test.csv | awk -F "," '{print $6}'
```

En este caso utilizamos dos comandos simultaneamente.
Grep para buscar la información de interés, y awk para imprimir únicamente la columna "Población". El **símbolo |** significa que el resultado de un programa es enviado al siguiente. De esta manera, muchos análisis pueden ser concatenados utilizando este símbolo.

A continuación guardaremos el resultado en un archivo llamado poblacion.txt, con el **símbolo >**.

```python
# Guardar el resultado en un archivo
!grep -- "-118.140000,33.880000" california_housing_test.csv | awk -F "," '{print $6}' > poblacion.txt
```

```python
# Se creó el archivo? qué tiene adentro?
!ls



```

Se creo el archivo poblacion.txt

```python
!cat poblacion.txt
```

Contiene la informacion de la columna 6, la poblacion de las filas que tenian las cordenadas de longitud y latitud "118.140000,33.880000"

## 5. Ejercicio extra

Google colab permite ejecutar código en diferentes lenguajes siempre y cuando esté bien especificado y escrito en la sintaxis correspondiente. El lenguaje por defecto es Python.

A continuación veremos como ejecutar diferentes lenguajes.

```python
# instalar R
!pip install rpy2
%load_ext rpy2.ipython
```

```python
# Ejecutar un código simple en R
%%R

suma <- 2 + 3
print("El resultado de la suma es:")
print(suma)
```

```python
# Ejecutar un código simple en Python

suma = 2 + 3
print("El resultado de la suma es:")
print(suma)
```

```python
# Ejecutar un código simple en Perl
%%perl

my $suma = 2 + 3;
print "El resultado de la suma es:\n";
print "$suma\n";
```

**Qué diferencias logra percibir?** 🤔



Se pueden observar varias diferencias entre los códigos ejecutados en R, Python y Perl, aunque todos realizan la misma operación básica.

En primer lugar, en R el resultado se muestra precedido por [1], lo cual indica la posición del elemento dentro de un vector. Esto refleja que R trabaja de manera nativa con estructuras vectoriales.

En Python, en cambio, la salida es más simple y directa, sin ningún indicador adicional, ya que imprime exactamente lo que se le indica con la función print().

Por otro lado, en Perl, aunque el resultado final es el mismo, se observa una sintaxis distinta, como el uso del símbolo $ para declarar variables y la inclusión explícita de \n para generar saltos de línea.

Además, otra diferencia importante es el uso de los llamados magic commands en entornos como Google Colab, donde se utilizan %%R y %%perl para ejecutar código en distintos lenguajes dentro del mismo notebook, mientras que Python es el lenguaje nativo y no requiere un prefijo especial.

En Google Colab es posible mezclar distintos lenguajes de programación dentro de un mismo cuaderno. Aunque las celdas se ejecutan por defecto en Python, se pueden usar comandos especiales (magics) para ejecutar otros lenguajes como R (%%R), Bash (! o %%bash), JavaScript (%%javascript), entre otros. Esto permite integrar análisis, automatización y visualización en un solo entorno interactivo.  

Los objetos/variables de python pueden ser llamadas dentro de otros lenguajes, utilizando el símbolo $ o {}
